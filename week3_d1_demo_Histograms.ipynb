#Creating Histograms

```python
# Take a look at the price column
df["SalePrice"].head()
```

```python
# Make a default histogram of the price column
df['SalePrice'].hist()
```

*Notice that the output started with <AxesSubplot:>

This means that panda's .hist returned an Axes.


```python
# saving the ax from .hist()
ax = df['SalePrice'].hist()
```

* The default settings do not create the optimal graph, but it is a good starting point. Graphing is an iterative process! Because there are so many aspects of creating visuals, this is an area where even experienced coders will need to Google and look at the documentation to obtain code to make all the desired tweaks. Don't try to memorize everything, but do get comfortable using your resources (often Google) to find what you need!
* One aspect of the above graph that we can change is removing scientific notation from the x-axis. To find out how to do this you could google "remove scientific notation from x-axis matplotlib" and you will find something similar to the code below:

```python
# Remove scientific notation
ax = df['SalePrice'].hist()
ax.ticklabel_format(style='plain');
```

* That worked to remove the scientific notation, but now our x labels are overlapping. We can address this by rotating the ticks using the ax.tick_params() method. We only want to rotate the x-axis, so we will specify "axis='x'" in addition to rotation=45 (to rotate the labels 45 degrees).


```python
# rotate your xticklabels
ax = df['SalePrice'].hist()
ax.tick_params(axis='x', rotation=45)
ax.ticklabel_format(style='plain');
```

* Next, we will spread the data out by using more bins by adding `bins=30` to .hist(). Note: the default number of bins is 10.

```python
# Change number of bins to 30
ax = df['SalePrice'].hist(bins = 30)
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

* We can also let Matplotlib and Pandas figure out the ideal number of bins by using bins='auto'.

* This will generally show a much more detailed/nuanced version of the histogram.

```python
# Change number of bins to auto
ax = df['SalePrice'].hist(bins = 'auto')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

* The code below also adds a black outline around each bin by setting the edgecolor to black by adding `edgecolor="black"` to `df.hist()`. Note that we also returned to using 30 bins here.

```python
# Change number of bins t0 30 and add edgecolor
ax = df['SalePrice'].hist(bins = 30, edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
```

* Finally, we will add titles and labels to the histogram

```python
# Add title and labels
ax = df['SalePrice'].hist(bins = 30, edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
ax.set_title('Distribution of Home Prices')
ax.set_xlabel('Price ($)')
ax.set_ylabel('Number of Houses')
```

# Histograms with Seaborn

##Seaborn's sns.histplot
* Seaborn has a sns.histplot function to create histograms.

* The best way to use Seaborn's plotting functions is to pass in the entire DataFrame as the data argument, and then just the column names for x or y.

* Seaborn will return a Matplotlib Axes that we can continue to work with, as we did earlier in the lesson using Panda's .hist().

```python
# Use seaborn's histplot function and save the ax
ax = sns.histplot(data=df, x="SalePrice")
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x', rotation = 45)
```

Notice that Seaborn did more work on our behalf:

* It added an xlabel (the column name)
* It added a ylabel ("Counts")
* It used bins='auto' by default

# Advantages of Seaborn's Histplot
Seaborn's plotting functions generally have many convenient parameters that make it easy to modify the resulting visualization without additional work.

For example, let's say we wanted to make bins for SalePrice that are $50,000 wide. The first bin would cover $0 to $50,000, the next would cover $50,000-$100,00, and so on.

If we tried to do this with Pandas/Matplotlib, we would have to manually create the list of bins and pass it in as the "bins" argument, but with Seaborn, we can simply specify the bin_width parameter.

```python
# Use binwidth arg to make histogram with $50K bins
ax = sns.histplot(data=df, x='SalePrice', binwidth=50_000)
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x', rotation = 45)
```

What if we also wanted to view the % of the data that's contained in each bin instead of the count?

* We can specify that we want stat='percent' instead of the default value, which is stat='count'

```python
# Use binwidth arg to make histogram with $50k bins and plot as % of total
ax = sns.histplot(data=df, x='SalePrice', binwidth=50_000, stat='percent')
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x', rotation = 45)
```

Our histogram shows us the distribution of home prices in this dataset. We see that most homes are priced at less than $500,000. There is a small percentage of homes that cost $0 - $50,000 (~ 1-2%), a larger number of homes were priced between $50,000 - $100,000 (~ 13%), but the most common price range is between $100,000 - $150,000 (~ 37%).
