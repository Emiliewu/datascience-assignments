```python
# Making a series from a dictionary
expenses = pd.Series({'brownie':0.25, 'cookie':0.5, 'cake':5, 'cupcake': 0.75},
                     name='Expenses')
expenses
```

```python
# saving our combined data as df
df = pd.concat([df, expenses], axis=1)
df
```

```python
# Calculate profit and define it as new column
df['Profit Per Item'] = df['Price'] - df['Expenses']
df
```

```python
# Calculate the profit per type of baked good and define as a column
df["Profits"] = df['Profit Per Item'] * df['Quantity Sold']
df
```

```python
# Find total profit
total_profit = df['Profits'].sum()
total_profit
```

```python
# This filepath will work if you saved your file in the recommend folder structure
import pandas as pd

fname = "/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/Our Bake Sale [Loading Data Vers] - Day 4 Sales.csv"
df = pd.read_csv(fname)
df.head()
```

```python
# Get the unique values and their count for the Student column
df["Student"].value_counts()
```

```python
# Use normalize = True to output a ratio
df["Student"].value_counts(normalize=True)
```

```python
# select quantity columns using .str.contains
columns = df.columns
filter_quantity_cols = columns[columns.str.contains('Quantity')]
filter_quantity_cols
```

```python
# Use the filter_quantity_cols index to limit the dataframe to Quantity columns
df[filter_quantity_cols]
```

```python
# Obtain the sum of the values in each column
df[filter_quantity_cols].sum()
```

```python
# Obtain the sum across rows
df[filter_quantity_cols].sum(axis=1)
```

```python
# Add a column to include the cumulative quantity
df['Cumulative Quantity (Day 4)'] = df[filter_quantity_cols].sum(axis=1)
df.head()
```

```python
# Sort the cumulative values from high to low
df['Cumulative Quantity (Day 4)'].sort_values(ascending=False)
```

```python
# Multiply two cumulative quantity by profit per item to get cumulative profit
df['Profit per Item'] = df['Price'] - df['Expense']
# Multiply two cumulative quantity by profit per item to get cumulative profit
df['Cumulative Profit (Day 4)'] = df['Cumulative Quantity (Day 4)']  * df['Profit per Item']
df.head()
```

```python
# Sort to find the most profitable item
df['Cumulative Profit (Day 4)'].sort_values(ascending=False)
```

```python
# groupby needs an aggregation functon
student_sums = df.groupby("Student").sum()
student_sums
```

```python
# Obtain the cumulative profit per student
quantity_by_student = df.groupby('Student')['Cumulative Quantity (Day 4)'].sum()
# Sort from highest to lowest profit
quantity_by_student.sort_values(ascending=False)
```

```python
# Obtain cumulative profit grouped by student
profit_by_student = df.groupby('Student')['Cumulative Profit (Day 4)'].sum()
# Sort from highest to lowest
profit_by_student.sort_values(ascending=False)
```

```python
# Total Profit for the club
total_profit = df['Cumulative Profit (Day 4)'].sum()
total_profit
```

```python
# How much more money do they need to earn?
1000-total_profit
```
