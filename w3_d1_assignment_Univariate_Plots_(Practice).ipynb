## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
```

## Load Data

```python
# Use the result of clean data assignment 
fname = "/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/titanic_clean_data.csv"
df = pd.read_csv(fname)
df.head()
```

## Identifying Our Categorical Features

```python
# Using select dtypes to get names of categorical columns
cat_cols = df.select_dtypes('object').columns
cat_cols
```

```python
# Checking how many categories in each column
df[cat_cols].nunique()
```

## Categorical Plots with Pandas

```python
# Saving the value counts
val_counts_survived = df['Survived'].value_counts()
val_counts_survived
```

* 1) Plot the counts for the number of passengers who Survived vs. those that died.

```python
# Plot value counts as a bar graph
ax = val_counts_survived.plot(kind='bar', stacked=True)
ax.set_ylabel("Count")
ax.set_xlabel("Survived Or Not")
ax.set_title("Counts of Passengers who Survived and Died");
ax.set_xticklabels(['Died', 'Survived'], rotation =0);
```

## Categorical Plots with Seaborn

```python
# create a count plot
ax = sns.countplot(data=df, x="Survived")
ax.set_ylabel("Count")
ax.set_xlabel("Survived Or Not")
ax.set_title("Counts of Passengers who Survived and Died");
ax.set_xticklabels(['Died', 'Survived'], rotation =0);
```

## Categorical Plots with Matplotlib

```python
# create a bar plot
fig, ax = plt.subplots()
data_refactored = df['Survived'].replace({0:'Died', 1:'Survived'})
counts = data_refactored.value_counts()

ax.bar(x=counts.index, height=counts.values)
ax.set_xlabel('Survived Or Not')
ax.set_ylabel('Count')
ax.set_title("Counts of Passengers who Survived and Died");
# ax.set_xticklabels(['Died', 'Survived'], rotation =0);
```

# 2) Plot a histogram of the distribution of Age.

## Plot a Histogram with Seaborn

```python
# use Seaborn to plot a histogram of the distribution of Ages:
ax = sns.histplot(data=df, x="Age")
ax.ticklabel_format(style="plain");
ax.tick_params(axis='x');
```

## Plot a Histogram with Panda

```python
# # use Pandas to plot a histogram of the distribution of Ages:
ax = df['Age'].hist(edgecolor='black')
ax.tick_params(axis='x')
ax.ticklabel_format(style='plain')
ax.set(title='Distribution of Ages', xlabel='Age', ylabel='Count');
```

## Plot a Histogram with Matplotlib

```python
# # use Matplotlib to plot a histogram of the distribution of Ages:
plt.hist(df['Age'], bins=20, color='blue', alpha=0.5, edgecolor='black')
plt.ticklabel_format(style='plain')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Distribution of Age");
```

## 3) Plot a boxplot for the Fare column.

```python
# use Seaborn to boxplot Fare:
ax = sns.boxplot(data=df, y="Fare")
```

```python
# Boxplots using matplotlib
fig, ax = plt.subplots()
data = df['Fare']
ax.boxplot(data)
ax.set_ylabel('Fare');
```
