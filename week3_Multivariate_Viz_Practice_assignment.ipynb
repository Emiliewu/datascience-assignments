```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 100)
```

```python
fname = '/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week03/Data/data-cleaning-core.csv'
df = pd.read_csv(fname)
df.head()
     
```

```python
#Plot the counts for the number of passengers by Sex, colored by if they Survived or not.
df.info()
```

```python
ax = sns.countplot(data=df, x='Sex', hue='Survived')
ax.set(title='# of Survivors by Sex');
```

â€‹Question: Which group had a greater number of survivors (Survived=1)?
Answer: Females had more survivors


```python
#Plot a barplot of the average number of "Parents or Children" for those that Survived vs. those that died. 
ax = sns.barplot(data=df, x='Survived', y='Parents or Children')
ax.set_title('Number of Parents or Children');
```

Question: which group had a higher average number of Parents or Children?

Survivors had a higher average number of parents or children.

```python
# Plot a corrleation heatmap of all of the numeric columns.
corr = df.corr(numeric_only=True)
ax = sns.heatmap(corr, annot=True, cmap='coolwarm');
ax.set(title='Correlation Heatmap');
```
